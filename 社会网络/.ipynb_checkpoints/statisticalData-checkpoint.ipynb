```python
# %load C:\Users\17702\Desktop\社会网络\statisticalData.py
import csv
from collections import defaultdict
import re
import spacy
import jieba
import jieba.analyse
from snownlp import SnowNLP

def chinese_sentiment_analysis(text):
    # 使用jieba进行分词
    seg_list = jieba.cut(text)
    # 使用SnowNLP进行情感分析
    sentiment_score = SnowNLP(" ".join(seg_list)).sentiments
    # 根据情感得分判断情感倾向
    if sentiment_score > 0.6:
        sentiment = '积极'
    elif sentiment_score < 0.4:
        sentiment = '消极'
    else:
        sentiment = '中性'
    return [sentiment,sentiment_score]

# 按照纪元划分时间段
years = ['危机纪元','威慑后','威慑纪元','广播纪元','掩体纪元','银河纪元','魔法时代','黄金年代']

# 人物名单和小说内容的文件路径
PEOPLE_FILE = '.\\数据\\people.txt'
for year in years:
    noval_file = 'txt/' + "years/" + year + '.txt'

    # 读取人物名单
    with open(PEOPLE_FILE, 'r', encoding='utf-8') as f:
        people = set(f.read().split('\n'))

    # 读取小说内容
    with open(noval_file, 'r', encoding='utf-8') as f:
        text = f.read()

    # 分段
    paragraphs = re.split(r'[。；！？:]', text)
    # print(paragraphs)

    # 用于存储所有段落的人物关系
    all_relations = defaultdict(lambda: defaultdict(int))  # 基于计数
    all_relations0 = defaultdict(lambda: defaultdict(float))# 基于情感分析

    # 分词
    custom_dict = {" 艾AA", "白Ice", "高Way", "A·J·霍普金斯", "1379号监听员"} #特殊名称
    for word in custom_dict:
        # 特殊名称加入分词库
        jieba.add_word(word)
    words = jieba.cut(text)

    # 统计每个人物的出现频次
    person_counter = defaultdict(int)
    for word in words:
        if word in people:
            person_counter[word] += 1

    # 遍历每个段落
    for paragraph in paragraphs:
        if paragraph.strip():  # 确保段落不为空
            # 统计
            pair_products = defaultdict(lambda: defaultdict(int)) # 基于计数
            pair_products0 = defaultdict(lambda: defaultdict(float)) # 基于情感分析
            for person1 in person_counter:
                for person2 in person_counter:
                    if person1 != person2:
                        pair_products[person1][person2] = person_counter[person1] * person_counter[person2]
                        if person1 in paragraph:
                            if person2 in paragraph:
                                # 进行情感分析
                                result = chinese_sentiment_analysis(paragraph)[0]
                                score = chinese_sentiment_analysis(paragraph)[1]
                                # print(f'分数：{score}, 情感分析结果： {result}')
                                pair_products0[person1][person2] = score

            # 将当前段落的人物关系添加到总关系中
            # 基于计数
            for person1, rel_dict in pair_products.items():
                for person2, product in rel_dict.items():
                    all_relations[person1][person2] += product
            # 基于情感分析
            for person1, rel_dict in pair_products0.items():
                for person2, score in rel_dict.items():
                    all_relations0[person1][person2] += score



    # 写入CSV文件
    with open('.\\数据\\person_relationships\\' + year + '.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # 写入标题行
        writer.writerow(['Person1', 'Person2', 'relationships'])
        # 打印最终的人物关系
        for person1, rel_dict in all_relations.items():
            for person2, strength in rel_dict.items():
                writer.writerow([person1, person2, strength])

    with open('.\\数据\\person_relationships0\\' + year + '.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # 写入标题行
        writer.writerow(['Person1', 'Person2', 'relationships'])
        # 打印最终的人物关系
        for person1, rel_dict in all_relations0.items():
            for person2, strength in rel_dict.items():
                writer.writerow([person1, person2, strength])


    with open('.\\数据\\new_person_counts\\' + year + '.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # 写入标题行
        writer.writerow(['Person', 'count'])
        # 打印最终的人物关系
        for person in person_counter:
                writer.writerow([person, person_counter[person]])

    print(year+":结果已写入到CSV文件中。")
```
